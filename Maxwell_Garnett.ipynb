##### General function

```python
def effective_permeance_mrf(up, um, con, r):
    """Calculate the effective relative permeance of a composite."""
    return um * ((up*con)/(r**2 * (1-r) * ((up/r) + um/(1-r))) + (1-con/r))
```

```python
%matplotlib inline
### THE GRAPH FOR THE R VALUES OF THE 4 MRF
X = np.linspace(0.5, 0.97)
Y = [effective_permeance_mrf(1000, 1, 0.33, r) for r in X]
Y1 = [effective_permeance_mrf(1000, 1, 0.28, r) for r in X]
Y2 = [effective_permeance_mrf(1000, 1, 0.40, r) for r in X]
Y3 = [effective_permeance_mrf(1000, 1, 0.24, r) for r in X]

fig, ax0 = plt.subplots(1, 1, figsize=(12, 8))
ax0.vlines(0.935, 0, 4.63, color='k', alpha=0.5)
ax0.vlines(0.938, 0, 6.25, color='k', alpha=0.5)
ax0.vlines(0.947, 0, 6.25, color='k', alpha=0.5)
ax0.vlines(0.952, 0, 9.06, color='k', alpha=0.5)
ax0.hlines(6.25, 0.5, 0.947, color='k', alpha=0.5)
ax0.hlines(4.63, 0.5, 0.935, color='k', alpha=0.5)
ax0.hlines(9.06, 0.5, 0.952, color='k', alpha=0.5)
ax0.hlines(2.50, 0.5, 0.952, color='k', alpha=0.5)
plt.plot(X, Y, color="midnightblue", label="MRF-132CG")
plt.plot(X, Y1, color="crimson", label="MAGNAFLO")
plt.plot(X, Y2, color="green", label="MRF-140CG")
plt.plot(X, Y3, color="cyan", label="DAMPRO")
plt.grid()
ax0.set_xlabel("r value [-]")
ax0.set_ylabel("effective relative permeability [-]")
ax0.legend()
plt.show()
```

### graph pour 01.12.2021

```python
def r_noP_H(H):
    fi_max = 0.775
    fi = 0.4
    H0 = 16_000
    return (1/3)*(fi_max + 2 - (fi_max-fi)*math.exp(-H/H0))
```

```python
def r_P_dP(dP):
    H = 50_000
    tau = 26_000
    b = 0.5
    k = b*dP**2/tau**2
    fi = 0.4
    mu_min = mu_r(0.4)
    return 1 - fi / (mu_r(r_noP_H(H)) - k*(mu_r(r_noP_H(H)) - mu_min) - 1)
```

```python
def mu_r(r):
    fi = 0.4
    return fi/(1-r) + 1
```

```python
H = np.linspace(0, 100_000, 1000)
r_noP = np.array(list(map(r_noP_H, H)))
mu_noP = np.array(list(map(mu_r, r_noP)))

dP = np.linspace(0, 26_000, 1000)
r_P = np.array(list(map(r_P_dP, dP)))
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])
ax.plot(H, r_noP, label="r value")
ax.legend()
ax.set_xlabel("H [A/m]")
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])
ax.plot(H, mu_noP, label="relative magnetic permeability")
ax.legend()
ax.set_xlabel("H [A/m]")
```

```python
fig, ax = plt.subplots(1, 1, figsize=[10, 6])
ax.plot(dP, r_P, label="r value")
ax.legend()
ax.set_xlabel("dP [Pa]")
```
